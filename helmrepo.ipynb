# Helm 레포

## 레포 관련 명령

- `helm repo list`
- `helm repo add `
- `helm repo remove`
- `helm repo update`

## 차트 검색 명령

- 레포내에서 차트를 검색할 때는 `helm search repo` 명령 사용
- `helm search repo`: 모든 레포의 전체 차트 목록 출력
- `helm search repo 키워드`: 레포 이름이나 차트 이름에서 특정 키워드 검색하여 출력
- `helm search repo 키워드 --versions`: 검색된 챠트의 모든 버전 목록 출력

## 차트 관련 명령

- `helm show chart 차트명 --version 버전`: 차트 내용 조회
- `helm pull 차트명 --version 버전`: 차트를 압축 파일로 다운로드
  - `--untar --untardir 디렉토리` 옵션으로 압축을 풀어서 다운로드 가능
